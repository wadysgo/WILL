```python
class Car:
    def __init__(self, model, brand, color):
        self.model = model
        self.brand = brand
        self.color = color

    def run(self):
        print("El auto " + self.model + " esta en movimiento.")

toyota = Car('Auris', 'Toyota', 'Cafe')
toyota.run()
```

```python
## A continuacion te dejamos otro ejemplo de una clase

class Person:

    def __init__(self, name, dni):
        self.name = name
        self.dni = dni

    def say_hello(self):
        print("Hola mi nombre es: " + self.name)

juan = Person("Juan", 1234)
juan.say_hello()

carlos = Person("Carlos", 9865)
carlos.say_hello()
print(carlos.dni)
carlos.dni = 890
print(carlos.dni)

## Para llamar atributos nos basta con . para acceder
## En cambio para los metodos necesitamos ()
## No lo olvides es muy importante
```

```python
class Robot:
    def __init__(self, name, color):
        self.name = name
        self.color = color
        self.age = 1
        self.batery = 8.5
        self.on = True

    def greet(self):
        print("hola mi nombre es: ", self.name)
        print("Mi edad es: ", self.age)
        print("Mi porcentaje de bateria es: ", self.batery)
        print("Mi estado de encendido es: ", self.on)


jarvis = Robot("jarvis", "rojo")
jarvis.greet()

ultron = Robot("ultron", "verde")
ultron.greet()

robotina = Robot("robotina", "azul")
robotina.greet()
```
